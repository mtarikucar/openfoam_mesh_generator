```python
import numpy as np
import matplotlib.pyplot as plt

N = 10

dx = 1.0 / (N - 1)
dy = 1.0 / (N - 1)

u = np.zeros((N, N))

u[:, 0] = 0
u[:, -1] = 0
u[0, :] = 0
u[-1, :] = np.sin(np.pi * np.linspace(0, 1, N))

convergence_criterion = 1e-6
error = 1
iterations = 0

while error > convergence_criterion:
    error = 0
    for i in range(1, N-1):
        for j in range(1, N-1):
            old_u = u[i, j]

            u[i, j] = ((u[i-1, j] + u[i+1, j]) * dy**2 +
                       (u[i, j-1] + u[i, j+1]) * dx**2) / (2 * (dx**2 + dy**2))
            error = max(error, np.abs(u[i, j] - old_u))
    iterations += 1

print(f"Gauss-Seidel metodu {iterations} iterasyonda yakınsadı.")

X, Y = np.meshgrid(np.linspace(0, 1, N), np.linspace(0, 1, N))
plt.contourf(X, Y, u, cmap='viridis')
plt.colorbar(label='Potansiyel u(x,y)')
plt.title('Laplace Denkleminin Sayısal Çözümü')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
